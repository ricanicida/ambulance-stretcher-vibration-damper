```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Natural frequencies

```python
df_omega = pd.read_csv('omega_n.csv')
```

```python
plt.figure(figsize=(12,6))
plt.plot('k', 'f1', data=df_omega[df_omega.mecanismo == 2])
plt.plot('k', 'f2', data=df_omega[df_omega.mecanismo == 2])
plt.grid()
plt.xlabel('k (N/m)')
plt.ylabel('Frequência (Hz)')
plt.title('Frequências naturais (Hz) da segunda arquitetura')
plt.show()
```

# Acceleration RMS

```python
df_rms_mec2 = pd.read_csv('RMS_mec2_vertical.csv')
```

```python
df_rms_mec3 = pd.read_csv('RMS_mec3_vertical.csv')
```

```python
df_rms_mec3_horizontal = pd.read_csv('RMS_mec3_horizontal.csv')
```

## RMS x frequency of the input signal of ocilation

```python
plt.figure(figsize=(12,6))
plt.plot('freq', 'rms_harmonico', data=df_rms_mec2[(df_rms_mec2.k == 2000)])
plt.plot('freq', 'rms_harmonico', data=df_rms_mec2[(df_rms_mec2.k == 3000)])
plt.plot('freq', 'rms_harmonico', data=df_rms_mec2[(df_rms_mec2.k == 4000)])
plt.grid()
plt.legend(['k = 2000 N/m', 'k = 3000 N/m', 'k = 4000 N/m'])
plt.xlabel('Frequência (Hz)')
plt.ylabel('RMS (m/s^2)')
plt.title('RMS da aceleração imposta a maca (m/s^2) pela frequência do movimento harmônico (Hz)')
plt.show()
```

## Mechanism 2 vertical oscillation

### Low frequencies f <= 0.5 Hz

```python
plt.figure(figsize=(12,6))
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 2000) & (df_rms_mec2.freq <= 0.5)])
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 3000) & (df_rms_mec2.freq <= 0.5)])
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 4000) & (df_rms_mec2.freq <= 0.5)])
plt.grid()
plt.legend(['k = 2000 N/m', 'k = 3000 N/m', 'k = 4000 N/m'])
plt.xlabel('Frequência (Hz)')
plt.ylabel('RMS (m/s^2)')
plt.title('RMS da aceleração vertical da maca (m/s^2) pela frequência do movimento harmônico (Hz) na segunda arquitetura')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 2000) & (df_rms_mec2.freq <= 0.3)])
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 3000) & (df_rms_mec2.freq <= 0.3)])
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 4000) & (df_rms_mec2.freq <= 0.3)])
plt.grid()
plt.legend(['k = 2000 N/m', 'k = 3000 N/m', 'k = 4000 N/m'])
plt.xlabel('Frequência (Hz)')
plt.ylabel('Transmissibilidade (adimensional)')
plt.title('Transmissibilidade da segunda arquitetura em função da frequência')
plt.show()
```

### f >= 0.3 Hz

```python
plt.figure(figsize=(12,6))
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 2000) & (df_rms_mec2.freq >= 0.3)])
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 3000) & (df_rms_mec2.freq >= 0.3)])
plt.plot('freq', 'rms', data=df_rms_mec2[(df_rms_mec2.k == 4000) & (df_rms_mec2.freq >= 0.3)])
plt.grid()
plt.legend(['k = 2000 N/m', 'k = 3000 N/m', 'k = 4000 N/m'])
plt.xlabel('Frequência (Hz)')
plt.ylabel('RMS (m/s^2)')
plt.title('RMS da aceleração vertical da maca (m/s^2) pela frequência do movimento harmônico (Hz) na segunda arquitetura')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 2000)])
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 3000)])
plt.plot('freq', 'transmissibilidade', data=df_rms_mec2[(df_rms_mec2.k == 4000)])
plt.grid()
plt.legend(['k = 2000 N/m', 'k = 3000 N/m', 'k = 4000 N/m'])
plt.xlabel('Frequência (Hz)')
plt.ylabel('Transmissibilidade (adimensional)')
plt.title('Transmissibilidade vertical da segunda arquitetura em função da frequência')
plt.show()
```
